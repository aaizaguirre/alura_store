# Introducción
El presente informe es un análisis sobre la situación actual de 04 tiendas pertenecientes a la cadena "Alura Store" con el fin de determinar la tienda menos eficiente. Para esto se tomaron en consideración 06 aspectos: 
- Ingreso total por cada tienda.
- Ventas por categoría.
- Valoración media por tienda.
- Productos más y menos vendidos.
- Valor del envío promedio por tienda.
- Análisis del desempeño geográfico.
Finalmente, se presenta una conclusión a partir de los datos analizados.

# Análisis

## Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

## 1. Análisis de facturación



```python
## Ingreso total por cada tienda
ingreso_total_tienda1 = tienda['Precio'].sum()
ingreso_total_tienda2 = tienda2['Precio'].sum()
ingreso_total_tienda3 = tienda3['Precio'].sum()
ingreso_total_tienda4 = tienda4['Precio'].sum()
print("Ingreso total tienda 1:", ingreso_total_tienda1)
print("Ingreso total tienda 2:", ingreso_total_tienda2)
print("Ingreso total tienda 3:", ingreso_total_tienda3)
print("Ingreso total tienda 4:", ingreso_total_tienda4)
```

```python
## Gráfico de barras del ingreso total por tienda
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [ingreso_total_tienda1, ingreso_total_tienda2, ingreso_total_tienda3, ingreso_total_tienda4]
plt.style.use('seaborn-v0_8')
plt.bar(tiendas, ingresos)
plt.title("Ingreso Total por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Ingreso total")
```

La tienda que genera un mayor ingreso es la **Tienda 01**; sin embargo, se observan ingresos muy similares.

## 2. Ventas por categoría

```python
productos_vendidos_categoria_tienda = tienda.groupby('Categoría del Producto').size().sort_values(ascending=False)
productos_vendidos_categoria_tienda2 = tienda2.groupby('Categoría del Producto').size().sort_values(ascending=False)
productos_vendidos_categoria_tienda3 = tienda3.groupby('Categoría del Producto').size().sort_values(ascending=False)
productos_vendidos_categoria_tienda4 = tienda4.groupby('Categoría del Producto').size().sort_values(ascending=False)
print(productos_vendidos_categoria_tienda)
print(productos_vendidos_categoria_tienda2)
print(productos_vendidos_categoria_tienda3)
print(productos_vendidos_categoria_tienda4)
```

```python
## Gráfico de barras de las ventas por categoría en cada tienda
productos_vendidos_categoria_tienda.plot(kind='bar', title='Ventas por categoría - Tienda 1')
productos_vendidos_categoria_tienda2.plot(kind='bar', title='Ventas por categoría - Tienda 2')
productos_vendidos_categoria_tienda3.plot(kind='bar', title='Ventas por categoría - Tienda 3')
productos_vendidos_categoria_tienda4.plot(kind='bar', title='Ventas por categoría - Tienda 4')
plt.xlabel("Categoría del Producto")
plt.ylabel("Número de productos vendidos")
```

Las categorías más populares en cada tienda son **Muebles** y **Electrónicos**. 

## 3. Calificación promedio de la tienda


```python
calificacion_tienda1 = tienda['Calificación'].mean()
calificacion_tienda2 = tienda2['Calificación'].mean()
calificacion_tienda3 = tienda3['Calificación'].mean()
calificacion_tienda4 = tienda4['Calificación'].mean()
print("Calificación promedio tienda 1:", round(calificacion_tienda1,2))
print("Calificación promedio tienda 2:", round(calificacion_tienda2,2))
print("Calificación promedio tienda 3:", round(calificacion_tienda3,2))
print("Calificación promedio tienda 4:", round(calificacion_tienda4,2))
```

```python
## Gráfico de barras para la calificación promedio por tienda
calificaciones = [calificacion_tienda1, calificacion_tienda2, calificacion_tienda3, calificacion_tienda4]
plt.bar(tiendas, calificaciones)
plt.title("Calificación Promedio por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Calificación promedio")
```

La tienda con mejor calificación promedio es la **Tienda 03**, mientras que la tienda con peor calificación promedio es la **Tienda 01**. Sin embargo, todas las tiendas poseen calificaciones similares y buenas. 

## 4. Productos más y menos vendidos

```python
productos_mas_vendidos_tienda1 = tienda['Producto'].value_counts().head(3)
productos_mas_vendidos_tienda2 = tienda2['Producto'].value_counts().head(3)
productos_mas_vendidos_tienda3 = tienda3['Producto'].value_counts().head(3)
productos_mas_vendidos_tienda4 = tienda4['Producto'].value_counts().head(3)
#print("Productos más vendidos tienda 1:\n", productos_mas_vendidos_tienda1)
#print("Productos más vendidos tienda 2:\n", productos_mas_vendidos_tienda2)
#print("Productos más vendidos tienda 3:\n", productos_mas_vendidos_tienda3)
#print("Productos más vendidos tienda 4:\n", productos_mas_vendidos_tienda4)
productos_menos_vendidos_tienda1 = tienda['Producto'].value_counts().tail(3)
productos_menos_vendidos_tienda2 = tienda2['Producto'].value_counts().tail(3)
productos_menos_vendidos_tienda3 = tienda3['Producto'].value_counts().tail(3)
productos_menos_vendidos_tienda4 = tienda4['Producto'].value_counts().tail(3)
print("Productos menos vendidos tienda 1:\n", productos_menos_vendidos_tienda1)
print("Productos menos vendidos tienda 2:\n", productos_menos_vendidos_tienda2)
print("Productos menos vendidos tienda 3:\n", productos_menos_vendidos_tienda3)
print("Productos menos vendidos tienda 4:\n", productos_menos_vendidos_tienda4)
```

Los productos más vendidos en la **Tienda 01** son: **Microondas, TV LED UHD 4K y Armario** y los menos vendidos son **Olla de presión, Auriculares con micrófono y Celular ABXY**.
Los productos más vendidos en la **Tienda 02** son **Iniciando en programación, Microondas y Batería** y los menos vendidos son **Mesa de comedor, Impresora y Juego de mesa**.
Los productos más vendidos en la **Tienda 03** son **Kit de bancas, Mesa de comedor y Cama King** y los menos vendidos son **Mochila, Microondas y Bloques de construcción**.
Los productos más vendidos en la **Tienda 04** son **Cama box, Cubertería y Dashboards con Power Bi** y los menos vendidos son **Guitarra acústica, Armario y Guitarra eléctrica**.

## 5. Envío promedio por tienda

```python
costo_envio_promedio_tienda1 = tienda['Costo de envío'].mean()
costo_envio_promedio_tienda2 = tienda2['Costo de envío'].mean()
costo_envio_promedio_tienda3 = tienda3['Costo de envío'].mean()
costo_envio_promedio_tienda4 = tienda4['Costo de envío'].mean()
print("Costo de envío promedio tienda 1:", round(costo_envio_promedio_tienda1,2))
print("Costo de envío promedio tienda 2:", round(costo_envio_promedio_tienda2,2))
print("Costo de envío promedio tienda 3:", round(costo_envio_promedio_tienda3,2))
print("Costo de envío promedio tienda 4:", round(costo_envio_promedio_tienda4,2))
```

```python
# Gráfico de líneas de envío promedio por tienda
costos_envio = [costo_envio_promedio_tienda1, costo_envio_promedio_tienda2, costo_envio_promedio_tienda3, costo_envio_promedio_tienda4]
plt.plot(tiendas, costos_envio, marker='o')
plt.title("Costo de envío promedio por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Costo de envío promedio")
```

El mayor costo de envío promedio se da en la **Tienda 01**, mientras que el menor costo de envío promedio se da en la **Tienda 04**.

# Resultados
La **Tienda 01** presenta un mayor ingreso total, mientras que la **Tienda 04** es la que menores ingresos genera. Así mismo, se puede observar que las categorías más vendidas son **muebles y electrónicos** en todas las tiendas; es decir, no existe una tienda que predominantemente venda más o menos productos de esta categoría respecto a las otras.
Con respecto a la calificación promedio por tienda, dato que representa la valoración que le dan los usuarios respecto a la experiencia de compra en cada local, se observa que la **Tienda 01** presenta el valor más bajo (3.98) y la **Tienda 03** el valor más alto (4.05).
El análisis de los productos más vendidos señala que la **Tienda 01** presenta la mayor venta de productos con mayor precio unitario, mientras que la **Tienda 04** presenta un mayor volumen de ventas de productos con menor precio unitario. Esto podría explicar el mayor valor de ingreso total de la **Tienda 01**.
Finalmente, podemos observar que los mayores costos de envío se dan en la **Tienda 01**, mientras que los menores costos se dan en la **Tienda 04**.




# Conclusión
Se observa que, si bien es cierto que la **Tienda 01** presenta los mayores ingresos, estos pueden deberse a una mayor venta de productos de mayor valor unitario. Sin embargo, estos productos pertenecen a categorías que también son las más vendidas en los otros locales; es decir, los márgenes de venta entre una tienda y otra son poco significativos, ya que no se deben a una alta proporción de venta o mayor cantidad de clientes. Así mismo, los costos de envío en esta tienda son los más altos con respecto a las demás, valor que también incide en los precios. Sin embargo, lo que si debe tenerse en consideración es la **baja valoración** que recibió la misma tienda por parte de los usuarios (3.98), lo que implicaría una mala experiencia de compra y que, a mediano o largo plazo, podría incidir en una disminución de ingresos. 
Debido a esto, se recomienda la venta de la **Tienda 01** con el fin de poder reinvertir el dinero en un nuevo emprendimiento.
